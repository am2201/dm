```python
from sklearn.datasets import load_iris, load_wine
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Loading the wine dataset
wine_x = load_wine().data
wine_features = load_wine().feature_names

# Loading the iris dataset
iris_x = load_iris().data
iris_features = load_iris().feature_names

print("Features for wine: ", wine_features)
print(pd.DataFrame(wine_x))

print("Features for iris: ", iris_features)
print(pd.DataFrame(iris_x))
```

```python
#Standardizing the datasets
wine_scaler = StandardScaler()
iris_scaler = StandardScaler()

wine_x_scaled = wine_scaler.fit_transform(wine_x)
iris_x_scaled = iris_scaler.fit_transform(iris_x)

print("Standardized Wine Data: \n", pd.DataFrame(wine_x_scaled))
print()
print("Standardized Iris Data: \n", iris_x_scaled)
```
